```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('artist_tracks.csv')
```

```python
# Descriptive statistics
print(df.describe())
```

```python
# Histograms
df['danceability'].hist()
plt.title('Distribution of Danceability')
plt.xlabel('Danceability')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Box plot
sns.boxplot(data=df, x='key', y='energy')
plt.title('Energy Distribution by Key')
plt.show()
```

```python
# Correlation maps per artist

plt.figure(figsize=(10, 8))
sns.heatmap(df[df['Artist'] == 'Bini'].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[df['Artist'] == 'Eddy Kenzo'].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
