```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Load the dataset
file_path = 'CONVENIENT_global_deaths.csv'  # Update this with the path to your CSV file
data = pd.read_csv(file_path)
```

```python
# Convert the 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Melt the dataframe to long format
data_long = pd.melt(data, id_vars=['Date'], var_name='Country/Region', value_name='Deaths')

# Aggregate global deaths by date
global_deaths = data_long.groupby('Date').sum().reset_index()
```

```python
# Plot global deaths over time using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Deaths', data=global_deaths)
plt.title('Global COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.show()
```

```python
# Analyze specific countries - replace 'Country1', 'Country2' with actual country names
countries = ['Philippines', 'Italy', 'China']  # Update or add countries as needed
country_data = data_long[data_long['Country/Region'].isin(countries)]
country_data = country_data.groupby(['Date', 'Country/Region']).sum().reset_index()

# Plot country-specific deaths over time using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Deaths', hue='Country/Region', data=country_data, marker="o")
plt.title('COVID-19 Deaths Over Time by Country')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.legend(title='Country')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter, FFMpegWriter
import seaborn as sns

sns.set(style="whitegrid")

from matplotlib.animation import writers
print(writers.list())

file_path = 'CONVENIENT_global_deaths.csv'  # Update this with the path to your CSV file
data = pd.read_csv(file_path)
data['Date'] = pd.to_datetime(data['Date'])
data_long = pd.melt(data, id_vars=['Date'], var_name='Country/Region', value_name='Deaths')

countries = ['Philippines', 'Italy', 'China']
filtered_data = data_long[data_long['Country/Region'].isin(countries)]
filtered_data = filtered_data.groupby(['Date', 'Country/Region']).sum().reset_index()

fig, ax = plt.subplots()

def update(num):
    ax.clear()
    sns.lineplot(x='Date', y='Deaths', hue='Country/Region', data=filtered_data[filtered_data['Country/Region'].isin(countries[:num+1])], marker="o", ax=ax)
    ax.set_title('COVID-19 Deaths Over Time by Country')
    ax.set_xlabel('Date')
    ax.set_ylabel('Number of Deaths')
    ax.legend(title='Country')
    ax.set_xlim([filtered_data['Date'].min(), filtered_data['Date'].max()])
    ax.set_ylim([0, filtered_data['Deaths'].max() + 1000])

ani = FuncAnimation(fig, update, frames=len(countries), repeat=False)
ani.save('covid_death_progression.gif', writer=PillowWriter(fps=0.5))
    
# if 'ffmpeg' in writers.list():
#     ani = FuncAnimation(fig, update, frames=len(countries), repeat=False)
#     ani.save('covid_death_progression.mp4', writer=FFMpegWriter(fps=0.5))
# else:
#     print("ffmpeg is not available, using Pillow to save as GIF")
#     ani = FuncAnimation(fig, update, frames=len(countries), repeat=False)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.animation import FuncAnimation
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Load the dataset
file_path = 'CONVENIENT_global_confirmed_cases.csv'  # Update this with the path to your CSV file
data = pd.read_csv(file_path)
data['Date'] = pd.to_datetime(data['Date'])

# Melt the dataframe to long format
data_long = pd.melt(data, id_vars=['Date'], var_name='Country/Region', value_name='Deaths')

# Group and sum data by date and country, then calculate cumulative sum
cumulative_data = data_long.groupby(['Country/Region', 'Date']).sum().groupby(level=0).cumsum().reset_index()

# Filter data for the selected countries
countries = ['Philippines', 'Italy', 'China']
cumulative_data = cumulative_data[cumulative_data['Country/Region'].isin(countries)]

# Plotting function for the animation
fig, ax = plt.subplots()

def update(num):
    ax.clear()
    ax.set_title('Cumulative COVID-19 Deaths Over Time')
    ax.set_xlabel('Date')
    ax.set_ylabel('Cumulative Number of Deaths')
    for country in countries:
        ax.plot(cumulative_data['Date'][cumulative_data['Country/Region'] == country][:num], 
                cumulative_data['Deaths'][cumulative_data['Country/Region'] == country][:num], label=f"{country}")
    ax.legend(title='Country')
    ax.grid(True)
    # Set up date formatting on the x-axis
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)  # Rotate dates for better visibility

# Create the animation
ani = FuncAnimation(fig, update, frames=len(cumulative_data['Date'].unique()), repeat=False)

# Save the animation as mp4
# ani.save('cumulative_deaths_animation.mp4', writer='ffmpeg', fps=2)
ani.save('cumulative_deaths_animation.gif', writer=PillowWriter(fps=6))

# Show the plot
# plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'CONVENIENT_global_confirmed_cases.csv'  # Update this with the path to your CSV file
data = pd.read_csv(file_path)
data['Date'] = pd.to_datetime(data['Date'])

# Melt the dataframe to long format
data_long = pd.melt(data, id_vars=['Date'], var_name='Country/Region', value_name='Deaths')

# Convert deaths to numeric, ensuring non-numeric values are handled
data_long['Deaths'] = pd.to_numeric(data_long['Deaths'], errors='coerce')

# Calculate daily deaths by country and date
data_daily = data_long.groupby(['Country/Region', 'Date']).sum().diff().reset_index()

# Removing the first day's NaN values that appear due to differencing
data_daily.dropna(subset=['Deaths'], inplace=True)

# Calculate the growth rate
data_daily['Growth Rate'] = data_daily.groupby('Country/Region')['Deaths'].pct_change().fillna(0) * 100

# Filter data for a few selected countries for clearer visualization
countries = ['Italy','China', 'Philippines']
data_growth = data_daily[data_daily['Country/Region'].isin(countries)]

# Plotting the growth rates
plt.figure(figsize=(12, 8))
sns.lineplot(x='Date', y='Growth Rate', hue='Country/Region', data=data_growth, marker="o")
plt.title('Daily Growth Rate of COVID-19 Deaths')
plt.xlabel('Date')
plt.ylabel('Growth Rate (%)')
plt.grid(True)
plt.legend(title='Country')
plt.show()
```
