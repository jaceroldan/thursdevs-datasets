```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv('duolingo/leaderboards.csv', parse_dates=['timestamp'])

# Display basic descriptive statistics
print(df['score'].describe())

# Visualization: Histogram of scores
sns.histplot(df['score'], kde=True)
plt.title('Distribution of Jaceeeee\'s Duolingo Leaderboard Scoring')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df.set_index('timestamp', inplace=True)  # Set timestamp as the index

# Plotting the time series of scores
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['score'], marker='o', linestyle='-')
plt.title('Time Series of Scores')
plt.xlabel('Date')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Summary statistics by tier
print(df.groupby('tier')['score'].describe())

# Visualization: Box Plot of Scores by Tier
plt.figure(figsize=(8, 5))
sns.boxplot(x='tier', y='score', data=df)
plt.title('Distribution of Scores by Tier')
plt.xlabel('Tier')
plt.ylabel('Score')
plt.show()

# Visualization: Bar Plot of Average Scores by Tier
plt.figure(figsize=(8, 5))
sns.barplot(x='tier', y='score', data=df, estimator=np.mean, ci=None)
plt.title('Average Scores by Tier')
plt.xlabel('Tier')
plt.ylabel('Average Score')
plt.show()
```
